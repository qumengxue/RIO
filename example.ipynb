```python
import cv2
from  matplotlib import pyplot as plt
import json

with open("RIO_common_all_thing.json") as f:
    data = json.load(f)
```

```python
this_id = 6666
coco_path = ""
dataType = 'val2014'
this_descrip = data[this_id]["expressions"]
this_box_list = data[this_id]["bbox_list"]
img_path = '{}/{}/COCO_{}_{}.jpg'.format(coco_path, dataType, dataType, str(data[this_id]["image_id"]).zfill(12))
```

```python
print(this_descrip)
this_img = cv2.imread(img_path)
for this_box in this_box_list:
    this_x1 = round(this_box[0])
    this_y1 = round(this_box[1])
    this_x2 = round(this_box[0] + this_box[2])
    this_y2 = round(this_box[1] + this_box[3])
    cv2.rectangle(this_img, (this_x1, this_y1), (this_x2, this_y2), (0,0,255), 2)
plt.imshow(this_img[:,:,::-1])
plt.axis('off')
plt.show()
```
